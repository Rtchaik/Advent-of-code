```python
def parse_data(my_file) -> dict:
    with open(my_file) as f:
        id_ranges, ingreds = f.read().split('\n\n')
        id_ranges = sorted([int(i) for i in ids.split('-')] for ids in id_ranges.split())
        new = []
        current = id_ranges[0]
        for ids in id_ranges[1:]:
            if ids[0]<=current[1]:
                if ids[1]>current[1]:
                    current[1]=ids[1]        
            else:
                new.append(current)
                current = ids
        new.append(current)
        return new, [int(ing) for ing in ingreds.split()]
```

```python
def part1(data: list) -> int:
    return sum(any(ids[0]<ing<=ids[1] for ids in data[0]) for ing in data[1])
```

```python
def part2(data: list) -> int:
   return sum(ids[1]-ids[0]+1 for ids in data)
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data[0]))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert part1(test1) == 3
assert part2(test1[0]) == 14
```
