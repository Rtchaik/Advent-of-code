```python
def parse_data(my_file) -> list:
    with open(my_file) as f:
        return [[int(num) for num in ids.split('-')] for ids in f.read().split(',')]
```

```python
def part1(data: list, part: int = 1) -> int:
    total = set()
    for ran in data:
        for div in ([2] if part else (set(range(2,len(str(ran[1]))//2+1))|{2}|set(range(len(str(ran[0])),len(str(ran[1]))+1)))-{1}):
            current = str(ran[0])
            while int(current)<=ran[1]:
                if len(current) % div == 0:
                    break
                current = f'1{"0"*len(current)}'
            else:
                continue
            current = current[:len(current)//div]
            
            while (num:=int(current*div
                           )) <= ran[1]:
                if num >= ran[0]:
                    total.add(num)
                current = str(int(current) + 1)
    return sum(total)
    
```

```python
def part2(data: list) -> int:
    return part1(data,0)
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data))
```
