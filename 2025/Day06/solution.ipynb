```python
from math import prod
import re
from itertools import chain
```

```python
def parse_data(my_file) -> list:
    with open(my_file) as f:
        return f.readlines()
```

```python
def cephalopods(data: list):
    return sum(
        (sum, prod)[prob[-1] == "*"](int(item) for item in prob[:-1]) for prob in data
    )
```

```python
def part1(data: list) -> int:
    return cephalopods(zip(*[line.split() for line in data]))
```

```python
def part2(data: list) -> int:
    data = re.findall(
        r"(?<![\\*+]).*?[\\*+]",
        "".join(chain.from_iterable(reversed(list(zip(*data)))))
    )
    return cephalopods(re.findall(r"\d+|[*+]", item) for item in data)
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert part1(test1) == 4277556
assert part2(test1) == 3263827
```
