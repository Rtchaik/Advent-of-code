```python
from itertools import product
```

```python
def parse_data(my_file) -> set:
    with open(my_file) as f:
        rolls = set()
        for y, line in enumerate(f.readlines()):
            for x, char in enumerate(line.strip()):
                if char == '@':
                    rolls.add(y+x*1j)
        return rolls
```

```python
def part1(data: set) -> set:
    adj8pos = list(map(sum,product([-1,0,1],[-1j,0j,1j])))
    total = set()
    for pos in data:
        if sum(pos+adj in data for adj in adj8pos) <= 4:
            total.add(pos)
    return total
```

```python
def part2(data: set) -> int:
    res = 0
    while (l:=len(s:=part1(data))):
        res += l
        data -= s
    return res
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', len(part1(data)))
print('Part 2: ', part2(data))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert len(part1(test1)) == 13
assert part2(test1) == 43
```
