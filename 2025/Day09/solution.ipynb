```python
from itertools import combinations
from math import prod
from shapely.geometry import Polygon, box
```

```python
def parse_data(my_file) -> list:
    with open(my_file) as f:
        return [tuple(int(coord) for coord in line.split(',')) for line in f.readlines()]
```

```python
def part1(data:list) -> int:
    return max(prod(abs(p[0]-p[1])+1 for p in zip(*pair)) for pair in combinations(data,2))
```

```python
def part2(data: list) -> int:
   data_copy = set(data)
   nodes = [data_copy.pop()]
   while data_copy:
      for item in data_copy:
         if any(item[x] == nodes[-1][x] for x in range(2)):
            nodes.append(item)
            data_copy.remove(item)
            break
   nodes = Polygon(nodes)
   area = 0
   for pair in combinations(data,2):
      (min_x,min_y),(max_x,max_y) = zip(*[sorted(p) for p in zip(*pair)])
      if nodes.contains(box(min_x,min_y,max_x,max_y)):
         area = max(area,(max_x-min_x+1)*(max_y-min_y+1))
   return area
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert part1(test1) == 50
assert part2(test1) == 24
```
