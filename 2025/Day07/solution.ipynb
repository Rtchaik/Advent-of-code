```python
from collections import defaultdict
from bisect import bisect
from functools import lru_cache
```

```python
def parse_data(my_file) -> tuple:
    with open(my_file) as f:
        start = f.readline().index('S')
        beams = defaultdict(list)
        for idx, line in enumerate(f.readlines(),1):
            for idx2, char in enumerate(line):
                if char == '^':
                    beams[idx2].append(idx)
    return start, beams
```

```python
def part1(start: int) -> int:
    global beams
    current = {(start, 0)}
    total = set()
    while current:
        new = set()
        for item in current:
            beam = beams[item[0]]
            splitter = bisect(beam,item[1])
            if splitter < len(beam):
                total.add((item[0],beam[splitter]))
                new |= {(item[0]+x,beam[splitter]) for x in (-1,1)}
        current = new
    return len(total)
```

```python
@lru_cache
def part2(item: tuple) -> int:
    global beams
    beam = beams[item[0]]
    splitter = bisect(beam,item[1])
    if splitter < len(beam):
        return part2((item[0]-1,beam[splitter])) + part2((item[0]+1,beam[splitter]))
    else:
        return 1
```

```python
%%time
start, beams = parse_data('raw.txt')
print('Part 1: ', part1(start))
print('Part 2: ', part2((start,0)))
```

```python
# tests
start, beams = parse_data("test_raw.txt")
assert part1(start) == 21
assert part2((start,0)) == 40
```
