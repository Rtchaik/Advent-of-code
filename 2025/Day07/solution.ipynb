```python
from collections import defaultdict
from bisect import bisect
from functools import lru_cache
```

```python
def parse_data(my_file) -> tuple:
    with open(my_file) as f:
        start = f.readline().index('S')
        beams = defaultdict(list)
        for idx, line in enumerate(f.readlines(),1):
            for idx2, char in enumerate(line):
                if char == '^':
                    beams[idx2].append(idx)
    return start, beams
```

```python
@lru_cache
def part2(item: tuple) -> int:
    global beams, splitters
    beam = beams[item[0]]
    splitter = bisect(beam,item[1])
    if splitter < len(beam):
        splitters.add((item[0],beam[splitter]))
        return sum(part2((item[0]+x,beam[splitter])) for x in (-1,1))
    else:
        return 1
```

```python
%%time
start, beams = parse_data('raw.txt')
splitters = set()
print('Part 2: ', part2((start,0)))
print('Part 1: ', len(splitters))
```

```python
# tests
start, beams = parse_data("test_raw.txt")
splitters = set()
assert part2((start,0)) == 40
assert len(splitters) == 21
```
