```python
from itertools import combinations
```

```python
def parse_data(my_file) -> list:
    with open(my_file) as f:
        return [line.strip() for line in f.readlines()]
```

```python
def part1(data: list, total: int = 2) -> int:
    return sum(max(map(lambda x: int(''.join(x)), combinations(bank,total))) for bank in data)
```

```python
def part2(data: list, total:int = 12) -> int:
    res = 0
    for bank in data:
        bat = []
        for pos in range(total,0,-1):
            bat.append(max(bank[:(None  if pos ==1 else 1-pos)]))
            bat_idx = bank.index(bat[-1])
            bank = bank[bat_idx+1:]
        res += int(''.join(bat))
    return res
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert part1(test1) == 357
assert part2(test1) == 3121910778619
```
