```python
from itertools import accumulate
```

```python
def parse_data(my_file) -> list:
    with open(my_file) as f:
        return [
            (-1 if item.startswith("L") else 1) * int(item[1:])
            for item in f.readlines()
        ]
```

```python
def part1(data: list) -> int:
    return list(accumulate(data, lambda x, y: (x + y) % 100, initial=50)).count(0)
```

```python
def part2(data: list) -> int:
    total = 0
    current = 50
    for i in data:
        if not current and i < 0:
            current = 100
        new = divmod(current + i, 100)
        if new[1] == 0 and i < 0:
            total += 1
        total += abs(new[0])
        current = new[1]
    return total
```

```python
%%time
data = parse_data('raw.txt')
print('Part 1: ', part1(data))
print('Part 2: ', part2(data))
```

```python
# tests
test1 = parse_data("test_raw.txt")
assert part1(test1) == 3
assert part2(test1) == 6
```
